```python
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data  = requests.get(url).text

soup = BeautifulSoup(html_data,"html5lib")

tesla_revenue = pd.DataFrame(columns=['Date', 'Revenue'])

for table in soup.find_all('table'):

    if ('Tesla Quarterly Revenue' in table.find('th').text):
        rows = table.find_all('tr')
        
        for row in rows:
            col = row.find_all('td')
            
            if col != []:
                date = col[0].text
                revenue = col[1].text.replace(',','').replace('$','')

                tesla_revenue = tesla_revenue.append({"Date":date, "Revenue":revenue}, ignore_index=True)

tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"")
tesla_revenue.dropna(inplace=True)

tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
tesla_revenue.tail()
```

```python
import matplotlib.pyplot as plt

# Supongamos que tesla_data tiene una columna 'Close' que representa el precio de cierre
def make_graph(data):
    plt.figure(figsize=(10, 5))  # Tamaño de la figura
    plt.plot(data.index, data['Close'], label='Precio de Cierre', color='blue')
    plt.title('Gráfico de Precios de Cierre de Tesla')  # Título del gráfico
    plt.xlabel('Fecha')  # Etiqueta del eje x
    plt.ylabel('Precio de Cierre (USD)')  # Etiqueta del eje y
    plt.legend()  # Muestra la leyenda
    plt.grid()  # Añade una cuadrícula
    plt.show()  # Muestra el gráfico

# Llama a la función para graficar los datos de Tesla
make_graph(tesla_data)
```
